```python
from gensim import corpora, models, similarities
from collections import defaultdict
import jieba
import io
stoplist = []
documents =[]
words=[]
with io.open('C:/Users/LCW/Desktop/stoplist.txt','r',encoding = 'utf-8') as f:
    for line in f.readlines():#一次讀一個停止詞並寫進stoplist
        stoplist.append(line.strip('\n'))

with io.open('C:/Users/LCW/Desktop/text/55555.txt','r',encoding = 'utf-8') as f:
    for line in f.readlines():#一次讀一篇文章(一篇文章一行)並寫進documents
        documents.append(line)
new=[[token for token in jieba.cut(doc) if token not in stoplist]for doc in documents]
#從documents一次讀一篇文章並斷詞，如果斷好的詞沒有出現在stoplist中，存在token，
#ex:[[第一篇文章斷好的詞],[第二篇文章斷好的詞],[第三篇文章斷好的詞],...]
frequency = defaultdict(int)#計算頻率網路上找到的...不會解釋
for text in new:#一次取一篇文章出來
    for token in text:#一次從一篇文章取一個詞出來
         frequency[token] += 1#詞的頻率+1
texts = [[token for token in text if frequency[token] > 1]
          for text in new]#一次取一篇文章出來，一次從一篇文章取一個詞出來，頻率>1留下來
dic = corpora.Dictionary(texts)#給每個詞一個id
corpus = [dic.doc2bow(text) for text in texts]#每篇文章出現哪些字以id表示
tfidf = models.TfidfModel(corpus)#作tfidf運算
corpus_tfidf = tfidf[corpus] 
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=6831)
#計算相似度 num_features print dic 可以看到unique token--->減一
lsi = models.LsiModel(corpus_tfidf, id2word=dic, num_topics=2)#套入lsi model num_topics是指要分幾個群
lsiout=lsi.print_topics(2)
print lsiout[0][1]
print lsiout[1][1]
corpus_lsi = lsi[corpus_tfidf]
for doc in corpus_lsi:
    print doc
```
